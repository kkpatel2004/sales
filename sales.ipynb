```python
df['ADDRESSLINE2'] = pd.to_numeric(df['ADDRESSLINE2'], errors='coerce')
df['ADDRESSLINE2'] = df['ADDRESSLINE2'].fillna(df['ADDRESSLINE2'].mean()) 
df.isnull().sum()
```

```python
df['STATE'] = pd.to_numeric(df['STATE'],errors='coerce') # # Convert the 'STATE' column to numeric, coercing non-numeric values to NaN
df['STATE'] = df['STATE'].fillna('Unknown')
df.isnull().sum()
```

```python
# Fill missing 'POSTALCODE' with '00000' 
df['POSTALCODE'] = df['POSTALCODE'].fillna('00000')
df.isnull().sum()
```

```python
# Fill missing 'TERRITORY' with 'Unknown' or the mode of the column
df['TERRITORY'] = df['TERRITORY'].fillna(df['TERRITORY'].mode()[0])  # or use 'Unknown'
df.isnull().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
# Plot histogram for 'SALES'
#plt.figure(figsize=(8, 6))
#df['SALES'].hist(bins=20, color='skyblue', edgecolor='black')
#plt.title('Distribution of Sales Amount', fontsize=16)
#plt.xlabel('Sales', fontsize=14)
#plt.ylabel('Frequency', fontsize=14)
#plt.grid(False)  # Turn off grid if preferred
#plt.show()


# Define conditions for coloring
low_sales = df['SALES'][df['SALES'] < 300]
medium_sales = df['SALES'][(df['SALES'] >= 300)]
high_sales = df['SALES'][df['SALES'] >= 400]

# Plot histogram with colors
plt.figure(figsize=(8, 6))

# Plot each segment with different colors
plt.hist(low_sales, bins=20, color='lightcoral', edgecolor='black', label='Low Sales (< 300)')
plt.hist(medium_sales, bins=20, color='gold', edgecolor='black', label='Medium Sales (300-399)')
plt.hist(high_sales, bins=20, color='limegreen', edgecolor='black', label='High Sales (>= 400)')

# Add titles and labels
plt.title('Distribution of Sales Amount', fontsize=16)
plt.xlabel('Sales', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.legend(fontsize=12)  # Add legend
plt.grid(False)  # Turn off grid if preferred
plt.show()
```

```python
# Assuming 'df' has a 'DATE' column and 'SALES' column
# Convert 'DATE' to datetime if it's not already in datetime format
df['ORDERDATE'] = pd.to_datetime(df['ORDERDATE'])

# Extract year and quarter from the 'DATE' column
df['Year'] = df['ORDERDATE'].dt.year
df['Quarter'] = df['ORDERDATE'].dt.to_period('Q')

# Group by Year and Quarter and calculate total sales
sales_by_year_quarter = df.groupby(['Year', 'Quarter'])['SALES'].sum().reset_index()

# Create a plot
plt.figure(figsize=(12, 6))

# Plotting the total sales by Year and Quarter
ax = sns.barplot(x='Quarter', y='SALES', hue='Year', data=sales_by_year_quarter, palette='RdYlBu')

# Set the title and labels
plt.title('Total Sales for Each Year and Quarter', fontsize=14)
plt.xlabel('Year and Quarter', fontsize=12)
plt.ylabel('Total Sales', fontsize=12)

for p in ax.patches:
    ax.annotate(f'{p.get_height():,.0f}', 
                (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha='center', va='center', 
                fontsize=10, color='black', 
                xytext=(0, 5), textcoords='offset points')

# Show the plot
plt.show()
```

```python
# Assuming df has 'DATE' and 'SALES' columns
# Convert 'DATE' to datetime if it's not already in datetime format
df['ORDERDATE'] = pd.to_datetime(df['ORDERDATE'])

# Extract Year and Month from the 'DATE' column
df['Year'] = df['ORDERDATE'].dt.year
df['Month'] = df['ORDERDATE'].dt.month

# Group by Year and Month and calculate total sales
sales_trend = df.groupby(['Year', 'Month'])['SALES'].sum().reset_index()

# Create a 'Date' column for easier plotting
sales_trend['Date'] = pd.to_datetime(sales_trend[['Year', 'Month']].assign(DAY=1))

# Plotting the trend of sales over time
plt.figure(figsize=(12, 6))
sns.lineplot(x='Date', y='SALES', data=sales_trend, marker='o', color='b')

# Set the title and labels
plt.title('Sales Trend Over Time (Months/Years)', fontsize=14)
plt.xlabel('Time', fontsize=12)
plt.ylabel('Total Sales', fontsize=12)

# Show the plot
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
customer_sales = df.groupby('CUSTOMERNAME')['SALES'].sum().reset_index()

# Sort the customers by total sales in descending order
customer_sales_sorted = customer_sales.sort_values(by='SALES', ascending=False)

# Display the top N high-value customers (e.g., top 10)
top_customers = customer_sales_sorted.head(10)
print(top_customers)
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='SALES', y='CUSTOMERNAME', data=top_customers, palette='Greens')

# Set title and labels
plt.title('Top 10 High-Value Customers Based on Total Sales', fontsize=14)
plt.xlabel('Sales', fontsize=12)
plt.ylabel('Customer Name', fontsize=12)

# Show the plot
plt.show()
```

```python
customer_sales = df.groupby('CUSTOMERNAME').agg(
    total_sales=('SALES', 'sum'),
    total_orders=('ORDERLINENUMBER', 'nunique')  # Count the unique number of orders
).reset_index()

# Calculate average order value (AOV) for each customer
customer_sales['AOV'] = customer_sales['total_sales'] / customer_sales['total_orders']

# Ensure that AOV was calculated correctly
print(customer_sales.head())
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='AOV', y='CUSTOMERNAME', data=customer_sales.head(), palette='twilight_shifted')

plt.title('Average Order Value (AOV) for Each Customer', fontsize=14)
plt.xlabel('Average Order Value (AOV)', fontsize=12)
plt.ylabel('Customer Name', fontsize=12)

plt.show()
```
